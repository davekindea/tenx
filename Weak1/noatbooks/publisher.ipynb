```python
from rake_nltk import Rake
import pandas as pd

# Initialize RAKE
r = Rake()

# Sample DataFrame
data = pd.DataFrame({
    'headline': [
        'Market hits record high.',
        'FDA approves new drug.',
        'Stock prices soar.',
        'New tech innovation released.',
        'Company reports record profits.',
        'Economic outlook improves.'
    ],'publisher': [
        'abc@example.com', 'def@example.com', 'abc@example.com',
        'ghi@example.org', 'abc@example.com', 'def@example.com'
    ]
})

# Function to extract keywords
def extract_keywords(text):
    r.extract_keywords_from_text(text)
    return r.get_ranked_phrases_with_scores()

# Apply keyword extraction
data['keywords'] = data['headline'].apply(extract_keywords)

# Display the DataFrame to verify
print(data)

# If you want to count keyword occurrences
keywords_by_publisher = data.groupby('publisher')['keywords'].apply(lambda x: x.explode().value_counts())
print("Keywords by Publisher:")
print(keywords_by_publisher)
```
